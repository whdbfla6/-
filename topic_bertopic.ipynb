```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
# from konlpy.tag import Mecab
from bertopic import BERTopic
```

```python
class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
    def __call__(self, sent):
        word_tokens = self.tagger.morphs(sent)
        result = [word for word in word_tokens if len(word) > 1]
        if result == None: result = ['문서']
        if len(result) == 0: result = ['문서']
        return result
```

```python
metadata = pd.read_json('processed_data/metadata.json')
metadata['sentence'] = metadata['title'] + ' ' + metadata['keyword_list'].apply(lambda x: ' '.join(x)) + ' ' + metadata['sub_title']
documents = metadata['sentence'].tolist()
```

```python
# custom_tokenizer = CustomTokenizer(Mecab())
# vectorizer = CountVectorizer(tokenizer=custom_tokenizer, max_features=3000)
vectorizer = CountVectorizer(max_features=3000)
```

```python
# # model_name = 'sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens'
# model_name = 'jhgan/ko-sroberta-multitask'
# nr_topics = 50
# top_n_words = 10
```

```python
def modeling(nr_topics,top_n_words,model_name = 'jhgan/ko-sroberta-multitask'):
  model = BERTopic(embedding_model=model_name,
                  vectorizer_model=vectorizer,
                  nr_topics=nr_topics,
                  top_n_words=top_n_words,
                  calculate_probabilities=False,
                  verbose=True)
  topics = model.fit_transform(documents)
  col_names = 'topic_num_{}'.format(nr_topics)
  metadata[col_names] = topics[0]
  return model
```

```python
# modeling(50,10)
model_100 = modeling(100,10)
model_20 = modeling(20,10)
```
